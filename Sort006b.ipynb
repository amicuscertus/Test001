```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import animation, rc
from IPython.display import HTML
 
n = 5
fig = plt.figure(figsize=(8, 6))
x = np.arange(n)
y = np.random.randint(0, 3 * n, n)
ims = []

def subset(work):
  print('len(work) = ', len(work))
  print('work input  = ', work)
  last = work[-1]
  work = work[0: len(work)-1]
  work = np.insert(work, np.searchsorted(work, last), last)
  print('work output = ', work)
  return work

def insertion_sort(arr):
  for i in range(1, len(arr) + 1):
    arr[0:i] = subset(arr[0:i])
    line, = plt.plot(x, arr, 'g', linewidth=5)
    ims.append([line])

insertion_sort(y)
anim = animation.ArtistAnimation(fig, ims, interval=200)
rc('animation', html='jshtml')
anim
```
