```
from skimage import data, io, filters
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np
```

```
image = io.imread('sweets.jpg')
io.imshow(image)
```

```
print(type(image))
print(image.shape)
```

```
reddish = image[:, :, 0] > 200
image_copy = np.copy(image)
image_copy[reddish] = [255, 0, 255]
io.imshow(image_copy)
```

```
io.imshow(reddish)
print(type(reddish))
print(reddish.shape)
```

```
celine = io.imread('celine-dion.jpg')
io.imshow(celine)
```

```
celine_gray = rgb2gray(celine)
io.imshow(celine_gray)
```

```
red_multiplier = [1, 0, 0]
yellow_multiplier = [1, 1, 0]

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax1.imshow(red_multiplier * celine)
ax2.imshow(yellow_multiplier * celine)
```
