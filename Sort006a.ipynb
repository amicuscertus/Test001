```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import animation, rc
from IPython.display import HTML
 
n = 30
fig = plt.figure(figsize=(8, 6))
x = np.arange(n)
y = np.random.randint(0, 3 * n, n)
ims = []

def subset(work):
  last = work[-1]
  for i in range(len(work)-1):
    if last < work[i]:
      for j in range(len(work) - 1 - i):
        work[len(work) - 1 - j] = work[len(work) - 1 - j - 1]
      work[i] = last
      break
  return

def insertion_sort(arr):
  for i in range(2, len(arr) + 1):
    subset(arr[0:i])
    line, = plt.plot(x, arr, 'r')
    ims.append([line])

insertion_sort(y)
anim = animation.ArtistAnimation(fig, ims, interval=200)
rc('animation', html='jshtml')
anim
```

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import animation, rc
from IPython.display import HTML
 
n = 5
fig = plt.figure(figsize=(8, 6))
x = np.arange(n)
y = np.random.randint(0, 3 * n, n)
print(y)
ims = []

def subset(work):
  print('before', work)
  last = work[-1]
  work = work[0: len(work)-1]
  print('work', work)
  print('last', last)
  work = np.insert(work, np.searchsorted(work, last), last)
  print('after', work)

def insertion_sort(arr):
  for i in range(2, len(arr) + 1):
    subset(arr[0:i])
    line, = plt.plot(x, arr, 'b')
    ims.append([line])

insertion_sort(y)
anim = animation.ArtistAnimation(fig, ims, interval=200)
rc('animation', html='jshtml')
anim
```
